```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
pokemon = pd.read_csv("./Pokemon.csv", index_col=0)
```

```python
plt.bar(pokemon['Type 1'], height=pokemon['Speed'])
plt.xticks(rotation=90)
plt.xlabel('Type')
plt.ylabel('Average Speed')
```

The task of user would be to compare the groups. 

### Not to use:
* Color as an encoding for Speed. Using saturation encoding will make it very hard to interpret the graph for users. Specifically, the user would assign red to low values and green to higher values, which may not be true in for the encoding scheme.
* Shape as an encoding for Speed. Speed in a quantitative variable, and from class, we know that shape is not effective for these

### Advantages 
* Encoding length of bar for speed is effective for compairing.

### Disadvantage
* Due to a large number of groups, the colors are repeated for different bars. This encoding could potentially for users. They could interpret similar colors to fall in a smiliar "group".

```python
legends = pokemon[pokemon.Legendary]
common = pokemon[pokemon.Legendary == False]
plt.scatter(legends['Type 1'], legends.Speed, label="Legendary", c='red', s=10)
plt.scatter(common['Type 1'], common.Speed, label="Common", c="blue", s=10)
plt.xticks(rotation=90)
plt.xlabel('Type')
plt.ylabel('Average Speed')
plt.legend(loc=1)
```

In this plot, the task would be identifying legendary pokemon stats.

### Not to use
* Using shape of point to encode type of pokemon. Although effective for categorial data, it will cause problems with such a large number of classes.
* Size of point for Legendary status. Encoding legendary pokemon with a bigger/smaller point will make it seem that they have a higer/lower speed. 

### Advanages
A scatterplot is very effective in helping with identification. The different color makes the points pop out, helping the user to quicky see the stats for those.

### Disadvantage
Using a scatterplot for categorial vs quantitative could be confusing.
